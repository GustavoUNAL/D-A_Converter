```python
from pydub import AudioSegment
import numpy as np

# Carga el archivo de audio
audio_file = "RollingintheDeep.mp3"
audio = AudioSegment.from_file(audio_file)

# Convierte el audio a muestras de 16 bits
samples = np.array(audio.get_array_of_samples())
```

```python
print(samples.shape)
print(samples.dtype)

# grafica las muestras
import matplotlib.pyplot as plt
# tamano de la grafica
plt.figure(figsize=(15, 5))
plt.plot(samples )

plt.xlim(0, len(samples))
plt.ylim(-2**15, 2**15)
plt.title('Rolling in the Deep')
plt.show()
```

## Reproducir audio 


```python
import numpy as np
import sounddevice as sd

# Parámetros para el tono
frecuencia = 440  # Frecuencia en Hz
duracion = 2  # Duración en segundos
fs = 44100  # Tasa de muestreo

# Generar la señal de audio
tiempos = np.arange(fs * duracion) / fs
senal = np.sin(2 * np.pi * frecuencia * tiempos)

# Reproducir la señal
sd.play(samples, fs)
sd.wait()  # Esperar a que termine la reproducción
```

```python
# grafica senal 
import matplotlib.pyplot as plt
# tamano de la grafica
plt.figure(figsize=(15, 5))
plt.plot(senal)
plt.xlim(0, len(senal))
plt.ylim(-1, 1)
plt.title('Tono de 440 Hz')
plt.show()
```
